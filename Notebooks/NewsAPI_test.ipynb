# 2nd approach (newsapi-python library)

```python
url = 'https://newsapi.org/v2/everything'
```

```python
api_key = 'c10b026da300469da0307a3a6f66c737'
```

```python
parameters = {
    'q' : 'Wildfire',
    'pageSize' : 100,
    'apiKey' : api_key,
    'language' : 'en',
    'from' : '2020-01-30'
}
```

```python
response = requests.get(url, params=parameters)
response_json = response.json()
```

```python
response_json
```

```python
response_json.keys()
```

```python
wildfire_articles = response_json['articles']
```

```python
wildfire_articles
```

```python
def get_articles(file):
    articles_result = []
    
    for i in range(len(file)):
        article_dict = {}
        article_dict['title'] = file[i]['title']
        article_dict['author'] = file[i]['author']
        article_dict['source'] = file[i]['source']
        article_dict['description'] = file[i]['description']
        article_dict['content'] = file[i]['content']
        article_dict['date'] = file[i]['publishedAt']
        article_dict['url'] = file[i]['url']
        article_dict['photo_url'] = file[i]['urlToImage']
        
        articles_result.append(article_dict)
    return articles_result
```

```python
get_articles(wildfire_articles)
```

```python
df_wildfire = pd.DataFrame(get_articles(wildfire_articles))
```
