# Construction d'un thésaurus baseline

L'objectif de ce notebook est l'élaboration d'un premier thésaurus "from scratch", qui pourra être par la suite amélioré.

**Pourquoi construire un thésaurus baseline?**

Dans un premier temps, nous avons récupéré le dataset 20NewsGroups qui contient environ 20 000 articles répartis en 20 catégories (allant de la médecine à l'informatique hardware en passant par l'Espace et l'athéïsme). 

Le but ultime sera donc de classifier ces articles grâce à notre thésaurus, et ainsi de tester l'hypothèse que l'on peut évaluer un thésaurus à partir d'une classification.

La première étape sera donc de récupérer un corpus de texte qui servira à la construction de ce thésaurus. On choisit de se concentrer sur l'Espace (spaceflight pour être encore plus précis). Nous allons donc sélectionner les articles les plus représentatifs de ce domaine, en récupérer leur contenu et construire notre premier thésaurus à partir de cela.

**#2 Extraction**

```python
def get_content(titles:list)->list:
    """For a given list of wikipedia page titles, 
    return a list of dict with the page title and the page content.
    :titles list of page titles
    :return list of content articles"""
    
    articles = []
    
    for title in tqdm(titles):
        article_dict = {}
        article_dict['title'] = title
        article_dict['content'] = wikipedia.page(title).content
        
        articles.append(article_dict)
        
    return articles
        
```

```python
articles = get_content(titles)
```

**#3 Prétraitement**

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from collections import Counter
```

```python
def get_dataframe(articles:list):
    """Return a clean DataFrame from a list of dict."""
    df = pd.DataFrame(articles)
    return df
```

```python
df_articles = get_dataframe(articles)
df_articles
```

```python
df_articles['content'][2]
```

```python
documents = df_articles['content'].tolist()
```

```python
stopwords = set(stopwords.words("english"))
```

```python
def get_lemma(word:str)->str:
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma
```

```python
def get_stem(word:str)->str:
    stemmer = SnowballStemmer(language='english')
    stem = stemmer.stem(word)
    return stem
```

```python
def parse(document:str):
    """For a given document, return text after applying tokenisation, normalization
    and stemming methods."""
    
    # Tokenisation
    tokens = document.split()
    # Normalisation (email addresses + ponctuation + caractères numériques + stopwords)
    tokens = [re.sub(r"\b[^\s]+@[^\s]+[.][^\s]+\b", "", token.lower()) for token in tokens]
    tokens = [re.sub("\W", "", token) for token in tokens]
    tokens = [token for token in tokens if token.isalpha() and token not in stopwords]
    #tokens = [get_lemma(token) for token in tokens]
    tokens = [get_stem(token) for token in tokens]
    
    return Counter(tokens)
```

```python
parsed_documents = [parse(doc) for doc in documents]
```

## Expérimentations

**Topic Modeling : test de la méthode LDA**

```python
# Imports

from gensim import corpora
import random
import pickle
import gensim
```

```python
def prepare_text(document:str):
    """For a given document, return text after applying tokenisation, normalization
    and stemming methods."""
    
    # Tokenisation
    tokens = document.split()
    
    # Normalisation (email addresses + ponctuation + caractères numériques + stopwords + lemmatization/stemming)
    tokens = [re.sub(r"\b[^\s]+@[^\s]+[.][^\s]+\b", "", token.lower()) for token in tokens]
    tokens = [re.sub("\W", "", token) for token in tokens]
    tokens = [token for token in tokens if token.isalpha() and token not in stopwords]
    #tokens = [get_lemma(token) for token in tokens]
    tokens = [get_stem(token) for token in tokens]
    
    return tokens
```

```python
# Build a list of all documents preprocessed

text_data = [prepare_text(doc) for doc in documents]
```

```python
# Build a bag-of-words with gensim

def get_bow(text_data:list):
    """Return a bag-of-words from a list of documents."""
    
    dictionary = corpora.Dictionary(text_data)
    corpus = [dictionary.doc2bow(text) for text in text_data]
    
    return dictionary, corpus

#dictionary = corpora.Dictionary(text_data)
#corpus = [dictionary.doc2bow(text) for text in text_data]
#pickle.dump(corpus, open('corpus.pkl', 'wb'))
#dictionary.save('dictionary.gensim')
```

```python
dictionary, corpus = get_bow(text_data)
```

```python
# Build a topic modeling model with LDA
# Return the n words most representative for the topic x

NUM_TOPICS = 4
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=5)
#ldamodel.save('model5.gensim')
topics = ldamodel.print_topics(num_words=11)
for topic in topics:
    print(topic)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
docs = ['this this this book',
        'this cat good',
        'cat good shit']
count_model = CountVectorizer(ngram_range=(1,1)) # default unigram model
X = count_model.fit_transform(docs)
#X[X > 0] = 1 # run this line if you don't want extra within-text cooccurence (see below)
Xc = (X.T * X) # this is co-occurrence matrix in sparse csr format
Xc.setdiag(0) # sometimes you want to fill same word cooccurence to 0
print(Xc.todense()) # print out matrix in dense format
```

```python
import scipy.sparse as sp
X[X > 0] = 1 # do this line first before computing cooccurrence
Xc = (X.T * X)
g = sp.diags(1./Xc.diagonal())
Xc_norm = g * Xc # normalized co-occurence matrix
```
